#### Importing the dataset

##### We shall be using the America Cancer Centre website dataset https://cancerstatisticscenter.cancer.org/#!/?module=BmVYeqHT

##### Importing the Incident rate dataset

```python
incident_rate_df = pd.read_excel("IncRate.xlsx", sheet_name =1,header = 6,index_col = 'State')
incident_rate_df.drop(incident_rate_df.index[:5],inplace=True)

# Displaying the first five records

incident_rate_df.head(5)
```

##### Importing the New Cases Estimates dataset

```python
new_case_estimates_df = pd.read_excel("NewCaseEstimates.xlsx", sheet_name =1,header = 6,index_col = 'State')
new_case_estimates_df.drop(new_case_estimates.index[:5],inplace=True)

# Displaying the first five records

new_case_estimates_df.head(5)
```

##### Importing the Death Estimates dataset

```python

death_estimates_df = pd.read_excel("DeathEstimates.xlsx", sheet_name =1,header = 6,index_col = 'State')
death_estimates_df.drop(death_estimates_df.index[:5],inplace=True)

# Displaying the first five records

death_estimates_df.head(5)

```
