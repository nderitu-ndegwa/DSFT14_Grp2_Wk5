#Data Analysis through Graphs

```python
plt.rcParams["figure.figsize"] = (100, 50)
xy = incidence_rate_state[['State','All cancer types combined / Both sexes combined']]
xy.plot.bar(x = 'State',fontsize='100')
plt.xlabel('State',fontsize='100')
plt.ylabel('Number of Incidences',fontsize='100')
plt.title('Number of Cancer Inccidences Per State',fontsize='100')
```

```python
plt.rcParams["figure.figsize"] = (100, 50)
xy = incidence_rate_race[['Cancer Type','American Indian and Alaska Native / Both sexes combined','Asian and Pacific Islander / Both sexes combined','Hispanic / Both sexes combined','Hispanic / Both sexes combined','Non-Hispanic black / Both sexes combined','Non-Hispanic white / Both sexes combined']]
xy.plot.bar(x = 'Cancer Type',fontsize='100')
plt.xlabel('Cancer Type',fontsize='100')
plt.ylabel('Number of Incidences',fontsize='100')
plt.title('Cancer Incidence Rates per Race',fontsize='100')
```

```python
plt.rcParams["figure.figsize"] = (100, 50)
xy = death_estimates[['State','All cancer types combined']]
xy.plot.bar(x = 'State',fontsize='100')
plt.xlabel('State',fontsize='100')
plt.ylabel('Number of Deaths',fontsize='100')
plt.title('Number of Deaths Caused by Cancer per State',fontsize='100')
```

```python
plt.rcParams["figure.figsize"] = (100, 50)
yx = new_case_estimates[['State','All cancer types combined']]
yx.plot.bar(x = 'State',fontsize='100')
plt.xlabel('State',fontsize='100')
plt.ylabel('Case Estimates',fontsize='100')
plt.title('New case Estimates per State',fontsize='100')
```
