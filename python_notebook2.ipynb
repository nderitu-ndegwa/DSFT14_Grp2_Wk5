**Since we have no data on certain types of Cancers we will Drop them for now as our study will be unfounded. We will go further to study the cancers that we actually have data on.**

```python
#Dropping columns with no Data
#.fillna('Unknown',inplace=True)
death_estimates_df.dropna(axis=1, how='all', inplace=True)
death_estimates_df.head()
```

```python
death_estimates_df.shape
```

**WE CAN STILL NOTE STRING ENTRIES ENTERED WHERE INT' VALUES SHOULD BE ENTERED, WE WILL CONVERT THE COLUMN VALUES INTO FLOAT DTYPE AND REMOVE THE STRINGS WE SEE ABOVE SUCH AS "NOT AVAILABLE" WHEN WE CAN JUST USE ZERO(0) INFACT LETS PREVIEW OUR DTYPES**

```python
death_estimates_df.dtypes
```

**NOTE ABOVE THAT THE COLUMNS DTYPE IS AN OBJECT THIS IS NOT PROPER ESPECIALLY IF WE WANT TO DO STATISTICAL COMPUTATIONS WITH THESE COLUMNS, WE WILL GO AHEAD AND CONVERT THEM**

```python
#We can do this by writing a function to replace all non_int values into an int like "0" or a a null value 
#thereafter we can fill the null values later on with either the mean, mode or any value that 
#computationally makes sense based on our dataset
#for now we will just replace these "Strings" with null values see below...
```

```python
#Check below..,you'll notice both string values and int values entered in the same columns, this is not good practice.
death_estimates_df.values
```

**CONVERTING OUR COLUMN VALUES TO NUMERIC**

```python
New_death_estimate_df=death_estimates_df.apply(pd.to_numeric, errors='coerce', axis=1)
New_death_estimate_df.head()
```

```python
#PROCEEDING TO DROP MORE NaN VALUES WHERE A COLUMN COMPLETELY HAS NO VALUE.
#NOTE THAT INITIALLY THESE COLUMNS WERE NOT DROPPED BECAUSE THEY WERE FILLED 
#WITH STRING VALUES THAT WE JUST DROPPED ABOVE. 

New_death_estimate_df.dropna(axis=1, how='all', inplace=True)
New_death_estimate_df.head(10)
```

**Looks a bit better now, we will now fill the remaining 'Nan' values. Remember the goal is to preserve as much as possible the original data values we have instead of dropping all the Nan values all at once, this is how we end up loosing all our data and its not good practice**

```python

#FILLING REMAINING 'NaN' VALUES 
#We will use a Backward fill and a Forward Fill simultaneously to ensure no Nan values are left

#FIRST THE BACKWARD FILL
New_death_estimate_df.fillna(method='bfill')
```

```python
#THEN WE DO THE FORWARD FILL
New_death_estimate_df.fillna(method='ffill')
```
